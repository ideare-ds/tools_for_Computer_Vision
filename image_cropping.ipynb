```python
import cv2
import os
import numpy as np
import ctypes
import glob
```

```python
def mouse_crop(event, x, y, flags, param):
    
    # grab references to the global variables
    global x_cross, y_cross, x_start, y_start, x_end, y_end, cropping, h, w, c, roi, font, F
        
    # if the left mouse button was DOWN, start RECORDING
    # (x, y) coordinates and indicate that cropping is being
    if event == cv2.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y
        cropping = True

    # Mouse is Moving
    elif event == cv2.EVENT_MOUSEMOVE:
        x_cross, y_cross = x, y
        if cropping == True:
            x_end, y_end = x, y

    # if the left mouse button was released
    elif event == cv2.EVENT_LBUTTONUP:
        # record the ending (x, y) coordinates
        x_end, y_end = x, y
        cropping = False # cropping is finished
        refPoint = [(int(x_start/F), int(y_start/F)), (int(x_end/F), int(y_end/F))]
        if len(refPoint) == 2: #when two points were found
            roi = oriImage[refPoint[0][1]:refPoint[1][1], refPoint[0][0]:refPoint[1][0]]
            cv2.imshow("Cropped", roi)
            
```

```python
def fitImage(image, W, H):
    height, width, depth = image.shape

    scaleWidth = float(W)/float(width)
    scaleHeight = float(H)/float(height)
    if scaleHeight > scaleWidth:
        imgScale = scaleWidth
    else:
        imgScale = scaleHeight

    newX,newY = image.shape[1]*imgScale, image.shape[0]*imgScale
    newimg = cv2.resize(image,(int(newX),int(newY)))
    return newimg, imgScale
```

```python
cropping = False
x_cross, y_cross, x_start, y_start, x_end, y_end = 0, 0, 0, 0, 0, 0
# im_name = 'IPS_2022-05-02.16.11.16.7190_frame7.jpg'
input_folder = 'dataset_oil_leaks/images/new_dataset/test_label/clean'
cropped_path = 'dataset_oil_leaks/cropped/new_dataset/test_label/clean'
# image = cv2.imread(input_folder+os.path.sep+im_name)
# oriImage = image.copy()
# h,w,c = image.shape

## get Screen Size
user32 = ctypes.windll.user32
W, H = user32.GetSystemMetrics(0), user32.GetSystemMetrics(1)

font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
input_images = os.listdir(input_folder)
input_images = [input_image for input_image in input_images if '.jpg' in input_image]
len(input_images)
```

```python
stop = False

for j, im in enumerate(input_images):
    image = cv2.imread(input_folder+os.path.sep+im)
    
    oriImage = image.copy()
    h,w,c = image.shape
    
#     image.reshape
    image, F = fitImage(image, W, H)
    cv2.namedWindow("image")
    cv2.setMouseCallback("image", mouse_crop)

    while True:

        i = image.copy()
        ii = image.copy()


        if not cropping:
            cv2.line(ii, (0, y_cross), (w, y_cross), (0, 0, 255), 2)    # horizontal line
            cv2.line(ii, (x_cross, 0), (x_cross, h), (0, 0, 255), 2)    # vertical line
            cv2.putText(ii,im,(10,80), font, 0.6, (0, 0, 255), 1, cv2.LINE_AA)
            cv2.putText(ii,f'image {j+1} of {len(input_images)}',(10,100), font, 0.6, (0, 0, 255), 1, cv2.LINE_AA)
            cv2.imshow("image", ii)
        elif cropping:
            cv2.rectangle(i, (x_start, y_start), (x_end, y_end), (255, 0, 0), 2)
            cv2.imshow("image", i)
            print(x_start, y_start, x_end, y_end)
        pressedKey = cv2.waitKey(10) & 0xFF
        
        if pressedKey == ord('q'):
            cv2.destroyAllWindows()
            break

        if pressedKey == ord('s'):
            cv2.imwrite(cropped_path+os.path.sep+im[:-4]+'_c.jpg',roi)
            
            x = (x_end+x_start)/(2*w*F)
            y = (y_end+y_start)/(2*h*F)
            width = (x_end-x_start)/(w*F)
            height = (y_end-y_start)/(h*F)
            writeLine = f"15 {x:f} {y:f} {width:f} {height:f}"
            
            labelName = input_folder+os.path.sep+im[:-4]+'.txt'
#             if os.path.isfile(labelName):
#                 print ("File exist")
#             else:
#                 print ("File not exist")
            with open(labelName, 'w') as f:
                f.write(writeLine)
            f.close()
                
                
        if pressedKey == ord('x'):
            stop = True
            cv2.destroyAllWindows()
            break
         
    if stop == True:
        break
```

### Label correction

```python
labeldirname = 'dataset_oil_leaks/images/new_dataset/dirty'

labeldir = os.listdir(labeldirname)
labeldir = [lab for lab in labeldir if 'txt' in lab]
```

```python
labeldir
```

```python
labelName = labeldirname
for lab in labeldir:
    with open(labeldirname+os.path.sep+lab) as f:
        line = f.readlines()
        f.close()
    line_split = line[0].split()
        
    line_split_correct = [15,]
    for i,coord in enumerate(line_split[1:]):
        line_split_correct.append(float(coord)*2.4)
            
    writeLine = f"15 {line_split_correct[1]:f} {line_split_correct[2]:f} {line_split_correct[3]:f} {line_split_correct[4]:f}"
        
    with open(labeldirname+os.path.sep+lab, 'w') as f:
        f.write(writeLine)
        f.close()
```
